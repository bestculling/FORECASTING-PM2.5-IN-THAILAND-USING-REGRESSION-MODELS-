```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
import folium
warnings.filterwarnings("ignore")
```

```python
df_station = pd.read_csv('data/datasets/Bangkok_Metropolitan_Region.csv')
df_station
```

```python
df_pm25 = pd.read_csv('data/air_pollution_2024_with_prep.csv')
df_pm25
```

```python
df_pm25 = df_pm25[['id', 'date', 'time', 'pm25']]
df_pm25
```

```python
filtered_pm25 = df_pm25[df_pm25['id'].isin(df_station['id'])]
filtered_pm25
```

```python
# print(filtered_pm25[filtered_pm25['id'] == '02t']['pm25'].mean())

# for i in df_station['id']:
#     print(i)
```

```python
data_coordinate = df_station['coordinate']
coordinates = []
for i in data_coordinate:
    latitude, longitude = i.split(",")
    coordinates.append([float(latitude), float(longitude)])
coordinates
```

```python
data_name = df_station['name']
for i in range(len(data_name)):
    coordinates[i].append(data_name[i])
    pm25 = filtered_pm25[filtered_pm25['id'] == df_station['id'][i]]['pm25'].mean()
    coordinates[i].append(round(pm25, 2))
    coordinates[i].append(df_station['id'][i])
coordinates
```

```python
map = folium.Map(location=[13.66791601805487, 100.60501734347488], zoom_start=12, tiles='OpenStreetMap')
for coordinate in coordinates:
    html = f"""
<div style="text-align: center">
  <h2>[{coordinate[4]}] {coordinate[2]}</h2>
  <h3>ค่าฝุ่น PM2.5 {coordinate[3]}</h3>
</div>
"""

    if coordinate[3] >= 75.1:
        color = "red"
    elif coordinate[3] >= 37.6:
        color = "orange"
    elif coordinate[3] >= 25.1:
        color = "yellow"
    elif coordinate[3] >= 15.1:
        color = "green"
    elif coordinate[3] >= 0:
        color = "blue"
    marker = folium.Marker(location=[coordinate[0], 
                                     coordinate[1]], 
                           popup = html,
                           tooltip="รายละเอียด",
                          icon=folium.Icon(color=color, icon="cloud"),
                          )
    marker.add_to(map)
map
```
