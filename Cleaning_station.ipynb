# ข้อมูลรายละเอียดสถานีตรวจวัดคุณภาพอากาศ
## http://air4thai.pcd.go.th/webV3/#/Report

```python
df_station = pd.read_excel('data/datasets/PM2.5(2023).xlsx', sheet_name='รายละเอียดจุดตรวจวัด')  
df_station
```

## เนื่องจากข้อมูลไม่ได้อยู่ในรูปแบบที่สามารถนำไปใช้งานได้ จึงต้องทำการ data cleaning

### ลบ 3 rows แรก เนื่องจากสถานนี 02T ควรเป็นสถานีแรกสุด

```python
df_station = df_station.drop([0, 1, 2]).reset_index(drop=True)
df_station
```

### ลบ column Unnamed: 0 และ รายละเอียดสถานีตรวจวัดคุณภาพอากาศ

```python
df_station = df_station.drop(columns=['Unnamed: 0', 'รายละเอียดสถานีตรวจวัดคุณภาพอากาศ '])
df_station
```

### เปลี่ยนชื่อ column ใหม่ให้ดังนี้
- Unnamed: 2 เป็น id
- Unnamed: 3 เป็น location
- Unnamed: 4 เป็น name

```python
df_station.columns = ['id', 'location', 'name']
df_station
```

### แปลง column id ให้เป็นตัวอักษรพิมพ์เล็ก

```python
df_station['id'] = df_station['id'].str.lower()
df_station['id']
```

### หา missing value

```python
df_station.isnull().sum()
```

```python
missing_ids = df_station[df_station['id'].isna()]
print(missing_ids)
```

### แทนที่ค่า NaN ใน column id ด้วย '70t'

```python
df_station['id'][69] = '70t'
df_station['id'][69]
```

```python
df_station
```

### Export df_station ออกมาเป็น csv ไฟล์  

```python
df_station.to_csv('data/df_station.csv', index=False)

if os.path.exists('data/df_station.csv'):
    print("export success")
else:
    print("export failed")
```
