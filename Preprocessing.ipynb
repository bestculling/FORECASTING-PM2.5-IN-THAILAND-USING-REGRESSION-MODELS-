```python
df = df[['id', 'date', 'time', 'pm25', 'pm10', 'o3', 'co', 'no2', 'so2']]
df
```

```python
fill_df = df.copy()
fill_df
```

```python
columns_to_fill = ['pm25', 'pm10', 'o3', 'co', 'no2', 'so2']
fill_df[columns_to_fill] = fill_df[columns_to_fill].fillna(fill_df[columns_to_fill].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set(style="whitegrid")

# Plot the data using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df.index, y='pm25', label='Original Data', color='blue')
sns.lineplot(data=fill_df, x=fill_df.index, y='pm25', label='Mean Data', linestyle='dotted', color='red')

# Set labels and title
plt.xlabel('Time')
plt.ylabel('PM2.5')
plt.title('Time Series of PM2.5 with Original Data (Blue) and Mean Data (Red)')

# Show legend
plt.legend()

# Show plot
plt.show()
```
